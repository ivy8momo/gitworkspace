```python
# 指定路径
import os
import pandas as pd
import numpy as np
from prettytable import PrettyTable

os.chdir('E:\Jupyterspace')
```

```python
# 读取知识库
inputfile = 'test03.xlsx'
symptom = pd.read_excel(inputfile,dtype={'rate_id':str,'score':str})
symptom.head()
```

```python
# 定制特定样例：重复出现属性，一个症状多个疾病，一个疾病多个症状
sample = symptom[symptom.no.isin(['107','106','105','349','122','126','66','242','350'])]
# 重复出现属性去重 symptom_attri_id
sample=sample.drop_duplicates(['symptom_attri_id'])
sample =  pd.DataFrame(sample,columns=['no'])
sample
```

```python
# 关联疾病、类型、等级
sample_ill = sample.merge(symptom)
sample_ill
```

```python
# num：按疾病、类型、等级分组，统计symptom_id去重数量
sympton_num = sample_ill.groupby(['relation_illn_id','type','class']).agg({'symptom_id': pd.Series.nunique})
sympton_num.columns = ['num']
sympton_num = sympton_num.reset_index()
sympton_num
```

```python
# 提取关联疾病列表
outcome = pd.DataFrame(sample_ill,columns=['relation_illn_id','relation_illn_name','sympton_name',
                                            'relation_type','relation_type_attri_value','rate','rate_id'])
outcome=outcome.drop_duplicates()
# ill_list = ill_list.set_index('relation_illn_id')

#增加症状数量num
outcome = outcome.merge(sympton_num)
outcome = outcome.set_index('relation_illn_id')
outcome
```

```python
rank_source = pd.DataFrame(outcome,columns=['type','class','num','rate_id'])
train_data = np.array(rank_source) #np.ndarray()
out_list = train_data.tolist() #list
out_concat = []
for i in rank_list:
    str1 = ''.join([str(e) for e in i])
    out_concat.append(str1)

out_sorc = pd.DataFrame(rank_concat,columns=['sorc'])
out_rank =pd.concat([outcome.reset_index(drop=True),out_sorc],axis=1)

out_rank['rank']=out_rank['sorc'].rank(ascending=0,method='dense')
out_rank = out_rank.sort_values(by = ['rank'],axis = 0,ascending = 1)

out_rank

```

```python
# 输出患者症状
symptom_list = out_rank['sympton_name']
symptom_list = symptom_list.drop_duplicates()
# 转化为list
train_data1 = np.array(symptom_list)#np.ndarray()
symptom_list=train_data1.tolist()#list

print('患者症状如下：')
s3 = ";".join([str(e) for e in symptom_list])
print(s3)
```

```python
# 输出疑似疾病与排名
ill_list = out_rank.loc[:,['rank','relation_illn_name']]
ill_list = ill_list.drop_duplicates()
ill_list

train_data2 = np.array(ill_list)  
ill_list=train_data2.tolist() 

print('疑似疾病如下：')
s4 = ";".join([str(e) for e in ill_list])
print(s4)
```

```python
# 输出排序原因
reason_list = out_rank.loc[:,['rank','relation_illn_name','sympton_name',
                              'relation_type','relation_type_attri_value','rate']]

train_data3 = np.array(reason_list)  
reason_list=train_data3.tolist() 


print('原因如下：')

x= PrettyTable(["排序","疑似疾病", "症状", "关系类型","关系类型属性值","症状权重"])
for i in reason_list:
    x.add_row(i)
print(x)
```

```python
filename = 'sample_out.txt'
f = open (filename,'a')
print('患者症状如下：',file = f)
s3 = ";".join([str(e) for e in symptom_list])
print(s3,file = f)
print('疑似疾病如下：',file = f)
s4 = ";".join([str(e) for e in ill_list])
print(s4,file = f)
print('原因如下：',file = f)
print(x,file = f)
f.close()
```
